### There are still a None value that we need to drop, and we should unify the records that point out to a private strip

### here is a function that turns a column into lower case

```python
def column_to_lowercase(data_frame, column_name):
    if column_name in data_frame.columns:
        data_frame[column_name] = data_frame[column_name].apply(lambda x: x.lower() if isinstance(x, str) else x)
    else:
        print(f"Column '{column_name}' not found in the DataFrame.")
```

```python
#transform the Airport name column into Lowercase
column_to_lowercase(df_airport_info, 'Airport.Name')

#Dropping the NONE Values
df_airport_info = df_airport_info[(df_airport_info['Airport.Name']!='none')]  

#unifying records that point out to a private strip
private_strip = ['private', 'private airstrip', 'private strip', 'pvt']
df_airport_info.loc[df_airport_info['Airport.Name'].isin(private_strip), 'Airport.Name'] = 'private strip'
```
